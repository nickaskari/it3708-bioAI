```python
import json
import matplotlib.pyplot as plt
```

```python
def plot_best_fitnesses_from_json(json_file_path):
 
    with open(json_file_path, 'r') as file:
        best_fitnesses = json.load(file)

    iterations = list(range(1, len(best_fitnesses) + 1))

    plt.figure(figsize=(18, 9))  
    plt.plot(iterations, best_fitnesses, marker='o', linestyle='--', color='b', linewidth=1, markersize=2)
    #plt.yscale('log')

    plt.xlim(0, max(iterations))  
    plt.ylim(0, 2000)  

    plt.title('Best Fitness Value Over Iterations')
    plt.yticks(range(0, 2001, 100))
    plt.xlabel('Iteration')
    plt.ylabel('Fitness')
    #plt.grid(False)
    plt.grid(True, which="both", ls="--")
    plt.show()



plot_best_fitnesses_from_json('bestFitnesses.json')
```
